```python
train = pd.read_csv("/kaggle/input/home-data-for-ml-course/train.csv")
test = pd.read_csv("/kaggle/input/home-data-for-ml-course/test.csv")
```

```python
train.head()
```

```python
from pycaret.regression import *

s = setup(train, target='SalePrice')
```

```python
best = compare_models()
```

```python
print(best)
```

```python
finalize_model(best)
```

```python
evaluate_model(best)
```

```python
predict_model(best)
```

```python
predictions = predict_model(best, test)
```

```python
submission = pd.DataFrame(np.array(predictions['prediction_label']), columns=['SalePrice'])
submission['id'] = submission.index
submission.head()
```

```python
submission[['id', 'SalePrice']].to_csv("submission_house_price.csv", index=False)
```
