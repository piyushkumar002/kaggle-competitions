```python
from sklearn import model_selection
```

```python
train['kfold'] = -1
```

```python
kf = model_selection.KFold(n_splits=5, shuffle=True, random_state=23)
for fold, (train_indices, valid_indices) in enumerate(kf.split(X=train)):
    train.loc[valid_indices, 'kfold'] = fold
```
